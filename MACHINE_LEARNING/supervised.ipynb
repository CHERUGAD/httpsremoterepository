```python
import pandas as pd

# Load existing CSV
df = pd.read_csv("titanic.csv")

# Add new columns with default or custom values
df["SibSp"] = [1,1,0,1,0,0,0,3,0,1]  # example values
df["Parch"] = [0,0,0,0,0,0,0,1,2,0]  # example values

# Save back to CSV
df.to_csv("titanic.csv", index=False)
df.info()
```

```python
df['Travelalone'] = np.where((df['SibSp'] + df['Parch']) > 0,0,1).astype('uint')
###condition: (df['sibsp'] + df['Parch']) > 0
###value_if_true: 0
###value_if_false: 1
#astype() ==Converts the data type of a column or array to another type.
#astype('uint') == Convert to unsigned integer type.
#np.where() == Returns elements chosen from two arrays element-wise.
#uint = Saves memory if your numbers are small.   Perfect for things like binary flags (e.g., 0 = no, 1 = yes), like your Travelalone column.
#	 uint8 : Size===	8 bits	     ,Value Range  ==0 to 255
```
