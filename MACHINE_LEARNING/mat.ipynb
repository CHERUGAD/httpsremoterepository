```python
import matplotlib.pyplot as plt
import numpy as np

# Define x and y
x = np.linspace(0, 10, 25)  # 25 values from 0 to 10
y = x ** 2 + 2              # y is a quadratic function of x

# Create the figure and axes
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # Position of the plot area
axes.plot(x, y, 'r')  # Plot x vs y with a red line

plt.show()
```

```python
fig,axes = plt.subplots(nrows= 1, ncols= 3)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #[ left, bottom, width, height ]

axes2 = fig.add_axes([0.1, 0.4, 0.3, 0.5])  # Position of the plot area
axes1.plot(x, y, 'r')
axes2.plot(x, y, 'g')
```

```python
fig = plt.figure(figsize=(5,6),dpi=100)
fig.add_subplot()
plt.plot(x, y, 'r')
```

```python

import matplotlib.pyplot as plt

# Custom x and y values (manually defined)
x = [1, 2, 3, 4, 5]
y = [10, 5, 8, 12, 7]

# Create plot
fig, axes = plt.subplots()
axes.set_title('Custom Data Plot')
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")
axes.plot(x, y, color='purple', marker='s', linestyle='-')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y1 = x ** 2
y2 = x ** 3

fig, axes = plt.subplots()
axes.plot(x, x**2)
axes.plot(x, x**3)
axes.legend(["y = x**2", "y = x**3"], loc=2)
plt.show()
```

```python
#In Matplotlib you can set other properties such as line color, transparency and more
fig,axes = plt.subplots(dpi=150)    #DPI  =  DOT PER INCH
axes.plot(x, x+1)
axes.plot(x ,x+2)
axes.plot(x, x+3)

```
